## Membership function for hot tempreature

## Universe of Discourse = 0 to 50 degree C

## Membership increases gradually = 25 to 45 degree c

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Temperature values (defines x axis values)

```python
temperature = np.linspace(0,50,100)
```

## Membership Function for Hot tempreature



```python
membership = []
for t in temperature: #if else defines the fuzzy logic rules
    if t <= 25:
        membership.append(0)
    elif 25 < t < 45:
        membership.append((t-25)/(45-25))
    else:
        membership.append(1)

```

## Plotting

```python
plt.plot(temperature,membership)
plt.xlabel("Temperature (degree C)")
plt.ylabel("Membership values")
plt.title("Membership function for hot temperature")
plt.ylim(0,1.1)
plt.grid(True)
plt.show()
```

## Interpretation

-->0°C to ~25°C → Membership = 0

-->Temperatures in this range are not considered hot.

~25°C to ~45°C → Linearly increasing (0 → 1)

-->The “hotness” gradually increases.

1)Around 30°C → slightly hot

2)Around 35°C → moderately hot

4)Around 40°C → very hot

≥ 45°C → Membership = 1

Temperatures at or above 45°C are considered fully hot.
