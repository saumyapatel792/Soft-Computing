```python
import numpy as np
import skfuzzy as fuzz #fuzzy logic functions 
import matplotlib.pyplot as plt
```

#Step 2: Defining universal of discourse

```python
import numpy as np

x = np.arange(0, 11, 1)   # define a range from 0 to 11 with a step of 1
print(x)
```

#Step 3:Creating a membership function

```python
low = fuzz.trimf(x,[0,0,5])
medium = fuzz.trimf(x,[0,5,10])
high = fuzz.trimf(x,[5,10,10])
```

#Step 4 : Ploting a membership function

```python
plt.figure()
plt.plot(x,low,label='Low')
plt.plot(x,medium,label='Medium')
plt.plot(x,high,label = 'High')

plt.title('Fuzzy Membership Functions')
plt.xlabel('Input Variables')
plt.ylabel('Membership Degree')

plt.legend()
plt.show()
```

Step 5: Calculation of fuzzy ratio for a value



```python
value = 4 # how much ratio of this temperature value belongs to a fuzzy set
low_degree = fuzz.interp_membership(x,low,value)
medium_degree = fuzz.interp_membership(x,medium,value)
high_degree = fuzz.interp_membership(x,high,value)

print("Membership Ratios:")
print("Low:", low_degree)
print("Medium:", medium_degree)
print("High :", high_degree)

```
